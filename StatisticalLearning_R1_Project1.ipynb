## 1. Steps for Importing and Cleaning of data


```python
#Store data from external csv to an initial datframe variable
df = pd.read_csv("./Laliga.csv")
```

```python
#Get some basic info on shape and data types etc
df.shape
df.info()
```

```python
#Also explore column-labels, row-indices, and a few rows of data.
df.head()
```

```python
#As the above shows, a presence of column-labels with names like "Unnamed:#" etc
#which isn't useful. While the row-index 0 shows the values, which sound more like
#column-labels. Hence the current column labels need to replaced, with the data from
#first-row data-values.Hence changing them by replacing with 0th row-index, as 
#given below
df.columns = df.iloc[0]
df.head()
```

```python
#Above step leaves us with extra row values, for 0th index which doesn't make sense any longer
#hence the whole data needs to be reindexed to start from 1st row instead.Storing the reindexed
#dataframe to a new dataframe variable: "laliga", which we plan to use in the later steps.
laliga =df.reindex(df.index.drop(0))
laliga.head()
```

```python
#The side effect of above is that we have a slightly "unconventional" datframe row-indices
#starting from 1st (instead of more traditional 0th start), hence reset-index is required
#with the extra intermediate "index" column dropped & saving it "inplace", in to the same
#variable, in single step: to "finally" get the data in the "desired usable format"!
laliga.reset_index(drop=True, inplace=True)
laliga.head()
```

### Q1. Read the data set and replace dashes with 0 to make sure, can perform arithmetic operations on the data

```python
#Replace all hyphens with values 0, and save it "inplace"(onto the laliga dataframe)
laliga.replace('-', 0, inplace=True)
laliga.head()
```

### Q2. Print all the teams which have started playing between 1930-1980.

```python
#Convert the Debut values to numeric-unsigned int type: 

#This requires us to first split years like 2002-03 etc. to be split at the hyphens, 
#and just the first part taken: i.e. 2002 for this given example.

#The second step is to convert the data-type to unsigned int, so that the series can be sorted
#on basis of the year etc.
laliga['Debut'] = pd.to_numeric(laliga['Debut'].map(lambda x:x.split('-', 1)[0]), downcast='unsigned')
```

```python
#Observe few rows
laliga.head()
```

```python
#Now the "Team" column can be filtered using the "Debut" as between
#year 1930-1980, as given below:
laliga.loc[(laliga.Debut >= 1930) & (laliga.Debut <= 1980), ["Team"]]
```

### Q3. Print the list of teams which came Top 5 in terms of points

```python
#Again convert the dataframe-"Points" column series, to an unsigned int type
#so that it can be sorted in either ascending or descending order of Points
#gained
laliga['Points'] = pd.to_numeric(laliga['Points'], downcast='unsigned')
```

```python
#Now sort the datframe by Points & store to a temp dataframe
#and display/print the 5 "Teams" thru the temp dataframe
#This ordering can be cross-validated thru the "Pos" column field
#as well.
teams_pointwise = laliga.sort_values(by=['Points'],ascending=False)
teams_pointwise.loc[:, ["Team"]].head(n=5)
```

### Q4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. Where the definition for: Goal_diff_count = GoalsFor - GoalsAgainst

```python
#Again, first convert the dataframe-"GoalsFor" column series, to an unsigned int type
laliga['GoalsFor'] = pd.to_numeric(laliga['GoalsFor'], downcast='unsigned')
```

```python
#Next, convert the dataframe-"GoalsAgainst" column series, to an unsigned int type
laliga['GoalsAgainst'] = pd.to_numeric(laliga['GoalsAgainst'], downcast='unsigned')
```

```python
#Define a func with name: “Goal_diff_count”, which returns the dataframe with an extra column
#named 'Goal_diff_count', which is a difference of passed dataframe columns 'GoalsFor' with
#'GoalsAgainst'.The given dataframe can "apply" this function to add  an extra column of "goal_diff_count"
#calculations.
def Goal_diff_count(df):
    df['Goal_diff_count'] = df['GoalsFor'] - df['GoalsAgainst']
    return (df)
```

```python
#The following line "applies" the above function and has to save to the 
#same dataframe to save the extra column "inplace".
laliga = laliga.apply(Goal_diff_count, axis=1)
```

```python
#observe the datframe to see the new column added(the "Goal_diff_count" one)
laliga.head()
```

```python
#This modified dataframe's "Goal_diff_count"series, can be used to return the team with 
#max series value, as given below ((in this case a value: +2807))
laliga.loc[(laliga.Goal_diff_count == laliga.Goal_diff_count.max()), ["Team"]]
```

```python
#Similarly, this modified dataframe's "Goal_diff_count"series can also be used, to return the team with 
#min series value, as given below (in this case a value: -525)
laliga.loc[(laliga.Goal_diff_count== laliga.Goal_diff_count.min()), ["Team"]]
```

### Q5. Create a new column with name “Winning Percent” and append it to the data set  Percentage of Winning = (GamesWon / GamesPlayed)*100. If there are any numerical error, replace it with 0%.Print the top 5 teams which has the highest Winning percentage

```python
#Again, first convert the dataframe-"GamesWon" column series, to an unsigned int type
laliga['GamesWon'] = pd.to_numeric(laliga['GamesWon'], downcast='unsigned')
```

```python
#Again, first convert the dataframe-"GamesPlayed" column series, to an unsigned int type
laliga['GamesPlayed'] = pd.to_numeric(laliga['GamesPlayed'], downcast='unsigned')
```

```python
#Set the float-format to two places after decimals
#And add an extra coumn named: 'Winning_Percent' which has:
# the Percentage of Winning = (GamesWon / GamesPlayed)*100 only if GamesPlayed is not equal to 0
#otherwise a divison by zero error can occur.
pd.options.display.float_format = '{:,.2f}'.format
laliga['Winning_Percent'] = np.where((laliga['GamesPlayed'] != 0), (laliga['GamesWon']/laliga['GamesPlayed'])*100, 0)
```

```python
#observe results
laliga.head()
```

```python
#Now sort the datframe by Winning_Percent & store to a temp dataframe
#and display/print the 5 "Teams" thru this new temp dataframe
#This ordering can be cross-validated thru the "Pos" column field
#as well.
teams_win_percent = laliga.sort_values(by=['Winning_Percent'],ascending=False)
teams_win_percent.loc[:, ["Team"]].head(n=5)
```

### Q6. Group teams based on their “Best position” and print the sum of their points for all positions.
Eg:---Best Position--------------------- Points----

         1                   25000

         2                    7000

```python
#Again, first convert the dataframe-"BestPosition" column series, to an unsigned int type
#The 'Points' column which needs to be summed up is already uint(due to earlier questions)
laliga['BestPosition'] = pd.to_numeric(laliga['BestPosition'], downcast='unsigned')
```

```python
#Groupby the dataframe based on the “Best position” and use the 'points' column
#to apply the "sum" agg-func on the same.The new multi-level column label is 
#also replaced as required by the Q6.
tmp_best_pos = laliga.groupby('BestPosition').agg({'Points':['sum']})
tmp_best_pos.columns = [ 'Points']
tmp_best_pos.head()
```
