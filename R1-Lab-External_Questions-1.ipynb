### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
#function to calulate range for any given "numerical" column
def range_cols(col_series):
    return (col_series.max()- col_series.min())
```

```python
#apply above func to all "numeric" columns only (i.e. except "Class" column)
df_range = Iris.loc[:, Iris.columns != "Class"].apply(range_cols).to_frame()
df_range.columns= ["Range"]
df_range
```

### Find out the no of male & female employees in the company

*Hint: Use can use a condition to find out the no of 1s and 0s in gender and sum it*

```python
#As the gender column has values 2's and 1's...here an assumption will have to be made, that lets say integer 2 represents males, 
#and int 1 repesents females.Asuming the same, we can filter the gender column for values 2 (i.e. male) and get a sum and divide 
#by 2 to get number of males... and similarly filter for value 1 and just sum it to find the number of females.
HR[["Gender"]].head()
```

```python
#Sum of the "gender" series, filtered by value 2 (i.e. assumption here 2 is used to encode males)  
sum_male_int_encoding = HR.loc[HR["Gender"]==2]["Gender"].sum()
```

```python
print("The number of males in the company: {}".format(int(sum_male_int_encoding /2)))
```

```python
#Sum of the "gender" series, filtered by value 1  (i.e. assumption here 1 is used to encode females)
sum_female_int_encoding = HR.loc[HR["Gender"]==1]["Gender"].sum()
```

```python
print("The number of females in the company: {}".format(sum_female_int_encoding))
```
